## AIML

### AIML1: PyTorch

```python
aiml1 = pd.read_csv('temp/aiml1/results.csv')
aiml1['COMMIT_DATE'] = pd.to_datetime(aiml1['COMMIT_DATE'], format='%Y-%m-%dT%H:%M:%SZ')
aiml1.head()
```

- Run the below code to check to get analystics with empty bot list
- Manually check if the the top committer seems to be a bot or not.
- If the top committer seems to be a bot, then add them to the bot list and re-run to get commit details with or without the bot.

```python
bots = ['pytorchmergebot@users.noreply.github.com', 'facebook-github-bot@users.noreply.github.com']
## Total shape
print("Total Shape")
print(aiml1.shape)
print("--")
## unique REASONs
print(aiml1['REASON'].value_counts())
print("--")
## top 5 max committer
print(" top 5 max committer")
print(aiml1['COMMITTER'].value_counts().head())
print("--")
## committers for valid commits
print("committers for valid commits")
print(aiml1[aiml1['REASON'] == 'valid']['COMMITTER'].value_counts())
print("--")
## Shape without webflow
print("Shape without webflow")
print(aiml1[aiml1['COMMITTER'] != 'noreply@github.com'].shape)
print("--")
## Shape without webflow and bot
print("Shape without webflow and bot")
print(aiml1[(aiml1['COMMITTER'] != 'noreply@github.com') & (~aiml1['COMMITTER'].isin(bots))].shape)
print("--")
## Shape without webflow and bot and reason counts
print("Without webflow and bot and reason counts")
print(aiml1[(aiml1['COMMITTER'] != 'noreply@github.com') & (~aiml1['COMMITTER'].isin(bots))]['REASON'].value_counts())
print("--")
## Shape without webflow and bot and reason counts before 28 april 2021
print("Without webflow and bot and reason counts before 28 april 2021")
print(aiml1[(aiml1['COMMIT_DATE'] < '2021-04-28') & 
            (aiml1['COMMITTER'] != 'noreply@github.com') & 
            (~aiml1['COMMITTER'].isin(bots))]['REASON'].value_counts())
print("--")
## Shape without webflow and bot and reason counts after 28 april 2021
print("Without webflow and bot and reason counts after 28 april 2021")
print(aiml1[(aiml1['COMMIT_DATE'] >= '2021-04-28') & 
            (aiml1['COMMITTER'] != 'noreply@github.com') & 
            (~aiml1['COMMITTER'].isin(bots))]['REASON'].value_counts())

```
